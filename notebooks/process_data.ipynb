# Process Tiny ImageNet

```python
import glob
import pandas as pd
```

```python
# df_train = pd.DataFrame(columns=['link','label'])
# df_train
f = open('data/tinyimagenet/processed/train.csv','w', encoding='UTF-8')
f.write("link,label\n")
```

```python
dir_train = glob.glob('data/tinyimagenet/raw/train/*')
for dr in dir_train:
    images = glob.glob(dr + '/images/*')
    for link in images:
        label = link.split('/')[-3]
        # print(link, label)
        f.write("{},{}\n".format(link, label))
```

```python
import pandas as pd
```

```python
df_train = pd.read_csv('data/tinyimagenet/processed/train.csv')
df_train
```

```python
len(set(df_train['label']))
```

```python
flabel = open('data/tinyimagenet/raw/wnids.txt','r')
```

```python
dict_labels = dict()
for idx, line in enumerate(flabel):
    if line[-1] =='\n':
        dict_labels[line[:-1]] = idx
    else:
        dict_labels[line] = idx
```

```python
len(dict_labels)
```

```python
dict_labels['n12267677']
```

```python
df_train['label'] = df_train['label'].apply(lambda x: dict_labels[x])
df_train
```

```python
len(set(df_train['label']))
```

```python
from sklearn.utils import shuffle
df_train = shuffle(df_train)
df_train
```

```python
df_train = df_train.reset_index(drop=True)
df_train
```

```python
df_train.to_csv('data/tinyimagenet/processed/train.csv', index=False)
```

```python
val_images = glob.glob('data/tinyimagenet/raw/val/images/*')
```

```python
fval = open('data/tinyimagenet/raw/val/val_annotations.txt', 'r')
val_labels = dict()
for line in fval:
    vals = line.split()
    name, label = vals[0], vals[1]
    val_labels[name] = label
    # print(name, label)
```

```python
f = open('data/tinyimagenet/processed/val.csv', 'w', encoding='UTF-8')
f.write("link,label\n")
# for link in val_images:
#     name = link.split('/')[-1]
#     label = val_labels[name]
#     f.write("{},{}\n".format(link,label))
```

```python
for link in val_images:
    name = link.split('/')[-1]
    label = val_labels[name]
    f.write("{},{}\n".format(link,label))
```
